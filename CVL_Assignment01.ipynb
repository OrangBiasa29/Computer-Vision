## Enhance Bright Image

Adjust the brightness of the original image to make it appear brighter by converting the image to float32 and add constant to make it brighter, then convert it back to uint8.


```python
img_float = img.astype(np.float32)
brightened_img = img_float + 70
brightened_img = np.clip(brightened_img, 0, 255).astype(np.uint8)
cv2_imshow(brightened_img)
```

## Enhance Dark Image

### Subtask:
Adjust the brightness of the original image to make it appear darker by converting the image to float32 and subtracting constant to make it brighter, then convert it back to uint8.

```python
img_float = img.astype(np.float32)
darkened_img = img_float - 50
darkened_img = np.clip(darkened_img, 0, 255).astype(np.uint8)
cv2_imshow(darkened_img)
```
